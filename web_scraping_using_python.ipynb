```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
x=input()

u = "https://www.flipkart.com/search?q="
url=u+x
response=requests.get(url)

soup=BeautifulSoup(response.content,'lxml')
names=[ ]
infos=[ ]
ratings=[ ]
prices=[ ]
discounts=[ ]

     
for item in soup.select('[data-id]'):
    try:
        name=(item.select('a img')[0]['alt'])
        info=(item.select('a')[0]['href']).split('/')[1]
        rating=(item.select('[id*=productRating]')[0].get_text().strip())
        p = item.find_all(text=re.compile('₹')) 
        price=(p[0])
        d = item.find_all(text=re.compile('off')) 
        discount=(d[0])
        column=[name,info,rating,price,discount]
        names.append(name)
        infos.append(info)
        ratings.append(rating)
        prices.append(price)
        discounts.append(discount)
        column=(name,price)
  
    except Exception as e:
#raise e
        b=0
import numpy as np
import pandas as pd

my_array = np.array([names,infos,prices,ratings,discounts])

df = pd.DataFrame(my_array).T
df.columns=["NAMES","INFORMATION","PRICES","RATINGS","DISCOUNT"]
```
